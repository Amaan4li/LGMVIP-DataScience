### First I have imported all the required libraries

```python
import numpy as np
import cv2
from PIL import Image
from IPython.display import display
```

### Here I am importing an Image to be converted to a Sketch

```python
from google.colab import files

# Upload the image file
uploaded = files.upload()
```

```python
import cv2
from google.colab.patches import cv2_imshow
img  = cv2.imread("cat.jpg")
img = cv2.resize(img,(640,1334))
rgb_img =  cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
display(Image.fromarray(rgb_img))
```

### Now I am converting that image to a greyscale

```python
greyscale_img =  cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(greyscale_img))
```

### Now I am turning that Greyscale into Negative

```python
neg_img = cv2.bitwise_not(greyscale_img)
display(Image.fromarray(neg_img))
```

```python
smooth_img = cv2.GaussianBlur(neg_img, (5,5), 0, 0)
display(Image.fromarray(smooth_img))
```

### And here I have conevrted the image to a Sketch

```python
sketch_img = 255 - smooth_img
sketch = cv2.divide(greyscale_img, sketch_img, scale=250.0)
display(Image.fromarray(sketch))
```

### Now displaying both the Images an an Image and a Sketch for Comparison

```python
rgb_img =  cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
display(Image.fromarray(rgb_img))
display(Image.fromarray(sketch))
```
